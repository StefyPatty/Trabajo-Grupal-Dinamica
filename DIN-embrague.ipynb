>Suponga que el sistema trabaja de manera horizontal. La fricción de las zapatas A contra el marco B hace funcional el embrague centrífugo y se transmite el torque de A hacia B. La fuerza centrífuga mantiene a las zapatas contra el marco B donde su coeficiente de fricción estático es de $\mu=0.8$. Cada zapata pesa 9.8[N] y el peso del resto de partes es despreciable. <br>
**a) Calcular la tensión inicial de los resortes (C) para que el contacto de las zapatas se haga cuando $\dot{\theta}\geq200[rpm]$<br>
b) Determinar el torque en el intervalo $100\geq\dot{\theta}\geq1500[rpm]$**

Realizando un diagrama de cuerpo libre y un diagrama de aceleraciones, se puede observar que la distancia $r=0,3m$ porque corresponde a la distacia desde el punto $O$ hasta el centro de masa. Por otra parte, la distancia $r_0=0,4m$ corresponde al brazo del torque correspondiente a la fuerza de rozamiento.<br>

Se coloca un sistema de referencia inercial en el centro de la carcasa. Desde una vista superior del sistema, se coloca un eje radial y transversal.<br>



Realizando un sumatorio de fuerzas en el eje radial se tiene:<br>
$\sum \vec{F_{e_{r}}}=m\vec{a_{r}}$<br>
$-N-T=m(\ddot{r}-r\dot{\theta}^2)$<br>
Como $\ddot{r}=0$, entonces:<br>
$N+T=mr\dot{\theta}^2$<br><br>
Para el literal a) se considera que $N=0N$ puesto que en el instante cuando $\dot{\theta}=200rev/min=20,94rad/s$ las zapatas recién estarían entrando en contacto con la carcasa. Entonces:<br>
$T=mr\ddot{\theta}^2=\frac{9,8}{9,8}*0,3*20,94^2$<br>
$T=131,55 N$<br><br>
Para el literal b), se puede observar que cuando $0rev/min\leq \dot{\theta}<200rev/min$ no existe torque, puesto que las pastillas no están en contacto con la carcasa. Se realiza la sumatoria de fuerzas en el eje radial para $200rev/min\leq\dot{\theta}\leq1500rev/min$. Entonces<br>
$\sum \vec{F_{e_{r}}}=m\vec{a_{r}}$<br>
$-T-N=m(\ddot{r}-r\dot{\theta}^2)$<br>
$T+N=r\dot{\theta}^2$<br>
$N=r\dot{\theta}^2-T$<br>
$\frac{fre}{\mu_{s}}=r\dot{\theta}^2-T$<br>
$fre=\mu_{s}(r\dot{\theta}^2-T)$<br><br>
Luego, para calcular el torque se aplica:<br>
$\vec{\tau}=\vec{r_{0}}\times\vec{fre}$<br>
$\vec{\tau}=r_{0}\vec{e_{r}}\times fre\vec{e_{\theta}}$<br>
$\vec{\tau}=r_{0}fre\vec{k}$<br>
$\tau=0,4*0,8*(0,3\dot{\theta}^2-131,55)$<br>
$\tau=0,096\dot{\theta}^2-42,096$<br><br>
Pero como sólo se realizó el análisis de una zapata, el torque total sería el doble, así:<br>
$\tau=0,192\dot{\theta}^2-84,192$<br><br>
Donde $\tau$ está en $Nm$ y $\dot{\theta}$ en $rad/s$<br>
Con esta función es posible hallar el torque generado cuando $200rev/min\leq\dot{\theta}\leq1500rev/min$, donde transformando a $rad/s$, se obtiene $20,944rad/s\leq\dot{\theta}\leq157,08rad/s$.<br>
De donde el torque en el intervalo $200rev/min\leq\dot{\theta}\leq1500rev/min$, es decir, en el intervalo de $20,944rad/s\leq\dot{\theta}\leq157,08rad/s$ es:
$0,029\leq\tau\leq4653,24$.<br>
En el intervalo de  $100rev/min\leq\dot{\theta}<200rev/min$ no existe torque, puesto que el contacto entre las zapatas y la carcasa comenzaría cuando $\dot{\theta}=200rev/min$

```python
import numpy as np #importar libreria matematica
import plotly.express as px #herramienta de plot
import plotly.graph_objects as go

R=300/1000 #[m]
Ro=400/1000 #[m]
us=0.8
rpm=200*((2*np.pi)/60) #[rad/s]
m=1
T=m*R*rpm*rpm

#Análisis
th=np.arange(20.944,157.08,1) #creando lista numérica
vth=np.array(th) #transformando a una matriz
vs=2*(Ro*us*R*vth*vth-Ro*us*T)
#Plot
fig=px.line(x=th,y=vs,labels={'x':'Velocidad angular [rad/s]','y':'Torque[Nm]'},title="Torque vs velocidad angular")
fig.show()
```
